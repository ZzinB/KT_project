```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector, make_column_transformer

def get_pipe(model, name: str) -> Pipeline:
    scaler = make_column_transformer(
        (StandardScaler(), make_column_selector(dtype_include=np.number)),
        remainder="passthrough"
    )

    pipe = Pipeline([
        ("scaler", scaler),
        (name, model)
    ])

    return pipe
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier, ExtraTreesClassifier, RandomForestClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

models = [
    ("svc", SVC(random_state=42)),
    ("dtree", DecisionTreeClassifier(random_state=42)),
    ("ada", AdaBoostClassifier(random_state=42)),
    ("etrees", ExtraTreesClassifier(random_state=42, n_jobs=-1)),
    ("gb", GradientBoostingClassifier(random_state=42)),
    ("rfc", RandomForestClassifier(random_state=42)),
    ("hist", HistGradientBoostingClassifier(random_state=42)),
    ("xgb", XGBClassifier(random_state=42, n_jobs=-1)),
    ("lgbm", LGBMClassifier(random_state=42, n_jobs=-1)),
    ("cat", CatBoostClassifier(random_state=42))
]

model_pipes = [(name, get_pipe(model, name)) for name, model in models]
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier, ExtraTreesClassifier, RandomForestClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

models = [
    ("xgb", XGBClassifier(random_state=42, n_jobs=-1)),
    ("lgbm", LGBMClassifier(random_state=42, n_jobs=-1)),
    ("cat", CatBoostClassifier(random_state=42))
]

model_pipes = [(name, get_pipe(model, name)) for name, model in models]
```

```python
import rich
from rich.table import Table
from sklearn.model_selection import cross_val_score
from tqdm.auto import tqdm

table = Table(title="점수 비교 테이블")
table.add_column("Model Name", justify="left", style="green")
table.add_column("Accuracy", justify="right")
voting_weights = []

for name, model in tqdm(model_pipes, leave=False):
    score = cross_val_score(model, x, y, scoring="accuracy", n_jobs=-1)
    score = np.mean(score)
    voting_weights.append(1 / score)
    table.add_row(name, f"{score:.4f}")

rich.print(table)
```

```python
from sklearn.ensemble import VotingClassifier, StackingClassifier

voting = VotingClassifier(model_pipes, weights=voting_weights, n_jobs=-1)
stacking = StackingClassifier(model_pipes, n_jobs=-1)

ensemble_table = Table(title="점수 비교 테이블")
ensemble_table.add_column("Model Name", justify="left", style="green")
ensemble_table.add_column("Accuracy", justify="right")

for name, model in tqdm([("voting", voting), ("stacking", stacking)]):
    score = cross_val_score(model, x, y, scoring="accuracy", n_jobs=-1)
    score = np.mean(score)
    ensemble_table.add_row(name, f"{score:.4f}")

rich.print(ensemble_table)
```
